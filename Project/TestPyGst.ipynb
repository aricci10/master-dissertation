```python
from __future__ import print_function #python 2 & 3 compatibility
import pygsti
from pygsti.extras import rb
import numpy as np
```

```python
nQubits = 1
qubit_labels = ['Q0'] 
gate_names = ['Gxpi2', 'Gxmpi2', 'Gypi2', 'Gympi2', 'Gzpi2', 'Gzmpi2','Gzpi','Gxpi','Gypi'] 
availability = {}
pspec = pygsti.obj.ProcessorSpec(nQubits, gate_names, availability=availability, 
                                 qubit_labels=qubit_labels)
```

```python
# The local error rate of every qubit
errorrate = 0.001

# Put these into a dictionary, mapping the qubit label to the error rate (which is qubit-indep here)
gate_errorrate_dict = {q : errorrate for q in qubit_labels}

# The error type : 'uniform' means locally uniform depolarization.
ptype = 'uniform'

# This creates this error model in the format needed for the simulator
errormodel = rb.simulate.create_locally_gate_independent_pauli_error_model(pspec, gate_errorrate_dict,
                                                                           ptype='uniform')

# The DRB lengths
lengths = [0,10,20,30,50,100,200,400]

# The number of circuits per length
k = 20

# The counts for each circuit. Below we use a basic stochastic-unravelling simulator,
# so the time taken is linear in the number of counts.
counts = 500
```

```python
filename = 'MySimulatedDRBData.txt'
rbdata = rb.simulate.rb_with_pauli_errors(pspec, errormodel, lengths, k, counts, 
                                          rbtype='CRB', filename=filename, verbosity=1)
```

```python
rbdata = rb.io.import_rb_summary_data('MySimulatedDRBData.txt')
rbresults = rb.analysis.std_practice_analysis(rbdata)

%matplotlib inline
rbresults.plot()
```

```python
def Rx(phi):
    resp=np.array([[np.cos(phi / 2), -1j * np.sin(phi / 2)],
       [-1j * np.sin(phi / 2), np.cos(phi / 2)]])
    
    return resp

def Ry(phi):
    
    resp= np.array([[np.cos(phi / 2), -np.sin(phi / 2)],
                   [np.sin(phi / 2), np.cos(phi / 2)]])
    return resp
```

```python
sigma_x=np.array([[0,1],[1,0]])

sigma_y=np.array([[0,-1j],[1j,0]])

Rypi2=Ry(np.pi/2)

Rympi2= Ry(-np.pi/2)

Rxpi2=Rx(np.pi/2)

Rxmpi2=Rx(-np.pi/2)

G2xpi=np.kron(sigma_x,sigma_x)

G2ypi=np.kron(sigma_y,sigma_y)

G2xpi2=np.kron(Rxpi2,Rxpi2)

G2xmpi2=np.kron(Rxmpi2,Rxmpi2)

G2ypi2=np.kron(Rypi2,Rypi2)

G2ympi2=np.kron(Rympi2,Rympi2)

```

```python
nQubits = 2
qubit_labels = ['Q0','Q1'] 

gate_names = ['Gzpi2', 'Gzmpi2','Gz','Gcnot'] 

nonstd_gate_unitaries = {'G2xpi':G2xpi,'G2ypi':G2ypi,'G2xpi2':G2xpi2,
                         'G2xmpi2':G2xmpi2,'G2ypi2':G2ypi2,'G2ympi2':G2ympi2}
availability = {}
pspec = pygsti.obj.ProcessorSpec(nQubits, gate_names, availability=availability, 
                                 qubit_labels=qubit_labels,nonstd_gate_unitaries=nonstd_gate_unitaries,verbosity=1)
```
