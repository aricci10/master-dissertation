```python
theta=np.pi

def Rx(phi):
    
    return np.matrix([[np.cos(phi/4),1j*np.sin(phi/4)],[1j*np.sin(phi/4),np.cos(phi/4)]])


def Rz(phi):
    
    return np.matrix([[np.cos(phi/2)+1j*np.sin(phi/2),0],[0,-1j*np.sin(phi/2)+np.cos(phi/2)]])
    
```

```python
initial=np.matrix([[1,0],[0,0]])
```

```python
def evaluatex (initial,phi):
    matrix= Rx(phi)@initial@Rx(phi).getH()
    
    w=(np.real(matrix[0,0])-np.real(matrix[1,1]))
    u=2*np.real(matrix[0,1])
    v=2*np.imag(matrix[0,1])
    
    return [u,v,w],matrix

def evaluatez (initial,phi):
    matrix= Rz(phi)@initial@Rz(phi).getH()
    
    w=(np.real(matrix[0,0])-np.real(matrix[1,1]))
    u=2*np.real(matrix[0,1])
    v=2*np.imag(matrix[0,1])
    
    return [u,v,w],matrix
```

```python
phis=np.linspace(0,np.pi/2,100)

results=[evaluatex(initial,phi)[0] for phi in phis]

phis=np.linspace(0,np.pi,100)

final1=evaluatex(initial,np.pi/2)[1]

results2=[evaluatez(final1,phi)[0] for phi in phis]

final2=evaluatez(final1,np.pi)[1]

phis=np.linspace(0,np.pi/2,100)

results3a=[evaluatex(final2,-phi)[0] for phi in phis]

results3b=[evaluatex(final1,-phi)[0] for phi in phis]

final3=evaluatex(final2,-np.pi/2)[1]

phis=np.linspace(0,np.pi,100)

results4=[evaluatez(final3,phi)[0] for phi in phis]
```
